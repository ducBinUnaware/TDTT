```python
#1
class Cylinder:
    def __init__(self,r : float,h : float):
        self.r=r
        self.h=h
        self.pi=3.14
    def S(self):
        return f"{2*self.pi*self.r*(self.r+self.h):.2f}"
    def V(self):
        return f"{self.pi*(self.r**2)*self.h:.2f}"
r,h=map(float,input().split())
c=Cylinder
cy=c(r,h)
print(cy.S(),cy.V())
```

```python
#2
class DATE:
    def __init__(self, year :int ,month : int, day : int):
        self.day=day
        self.month=month
        self.year=year
        self.days= ([0,31,29,31,30,31,30,31,31,30,31,30,31] if (year%4==0 and year%100!=0) or year%400==0 else [0,31,28,31,30,31,30,31,31,30,31,30,31])
    def tomorow(self):
        if self.year<0:
            print("INVALID")
            return
        if not 1<=self.month<=12:
            return("INVALID")
        if not 1<=self.day<=self.days[self.month]:
            return("INVALID")
        if (self.day==31 and self.month==12):
            return(f"01/01/{self.year+1}")
        if (self.day==self.days[self.month]):
            return (f"01/{self.month+1}/{self.year}")
        else:
            return (f"{self.day+1}/{self.month}/{self.year}")
d,m,y=map(int,input().split())
today=DATE(y,m,d)
print(today.tomorow())         
```

```python
#3
class Point:
    def __init__(self,x : float, y : float):
        self.x=x
        self.y=y
    def pos(self):
        if (self.x==0 and self.y==0): return("Diem M la goc toa do")
        elif (self.x==0 and self.y!=0): return("Diem M nam tren truc tung")
        elif(self.x!=0 and self.y==0): return("Diem M nam tren truc hoanh")
        else: return("")
    def dist(self,x1,y1):
        return f"{((x1-self.x)**2 + (y1-self.y)**2)**0.5:.2f}" 
x,y=map(float,input().split())
M=Point(x,y)
print(M.pos())
print(M.dist(0,0))
```

```python
#4
class MyCalculator:
    def __init__(self,x : float,y : float):
        self.x=x
        self.y=y
    def add(self):
        return (self.x+self.y)
    def sub(self):
        return (self.x-self.y)
    def product(self):
        return (self.x*self.y)
    def divide(self):
        if self.y==0: return("INVALID")
        return self.x/self.y
    def mod(self):
        if self.y==0: return("INVALID")
        return self.x%self.y
    def change(self,newx,newy):
        self.x=newx
        self.y=newy
def main():
    x,y=map(float,input("Enter two numbers: ").split())
    myNums=MyCalculator(x,y)
    while(1):
        print("---------------------------------")
        print("|                               |")       
        print("|            1.ADD              |")       
        print("|            2.SUBTRACT         |")       
        print("|            3.PRODUCT          |")       
        print("|            4.DIVIDE           |")       
        print("|            5.MOD              |")       
        print("|            6.CHANGE           |")       
        print("|            0.QUIT             |")       
        print("|                               |")       
        print("---------------------------------")
        try:
            choice=int(input("Choose one number(0->6): "))
        except:
            choice=int(input("INVALID!! Please choose again: "))
        while(not 0<=choice<=6): choice=int(input("INVALID!! Please choose again: ")) 
        else:
            if (choice==1): print(f"Result: {myNums.add()}\n")
            elif(choice==2): print(f"Result: {myNums.sub()}\n")
            elif(choice==3): print(f"Result: {myNums.product()}\n")
            elif(choice==4): print(f"Result: {myNums.divide()}\n")
            elif(choice==5): print(f"Result: {myNums.mod()}\n")
            elif(choice==6): 
                a,b=map(float,input("Please choose two numbers to change: ").split())
                myNums.change(a,b)
                print("New numbers successful change!!\n")
            else:
                s=input("Quit the program?(yes/Yes/YES to quit): ")
                if s=="yes" or s=="Yes" or s=="YES":
                    print("Program quited!!!")
                    break
main()

        
```
