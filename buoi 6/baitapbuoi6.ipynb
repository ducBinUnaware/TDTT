```python
#w6a1
def difnum(l):
    res=list()
    dt=dict()
    for i in range (len(l)):
        if dt.get(l[i]):
            continue
        dt[l[i]]=i
        res.append(l[i])
    return res
l=list(map(int,input().split()))
print(difnum(l))
```

```python
#w6a2
def listsum(nums):
    dp=[0]*len(nums)
    dp[0]=nums[0]
    for i in range (1,len(nums)):
        dp[i]=dp[i-1]+nums[i]
    return dp
nums=list(map(int,input().split()))
print(listsum(nums))

```

```python
#w6a3
def spin(nums,k):
    n=len(nums)
    if k>=n : return nums
    l=list()
    for i in range (k,n):
        l.append(nums[i])
    for i in range (k):
        l.append(nums[i])
    return tuple(l)
nums=tuple(map(int,input().split()))
k=int(input())
print(spin(nums,k))
    
```

```python
#w6a4
def stod(s):
    s=s+' '
    n=len(s)
    dt=dict()
    j=0
    k=0
    for i in range(n):
        if(s[i]==':'):
            if (not dt.get(s[j:i])):
                l=[]
                k=i
            else:
                l=dt[s[j:i]]
                k=i
        elif(s[i]==' '):
            l.append(s[k+1:i])
            dt[s[j:k]]=l
            j=i+1
    return dt
s=input()
print(stod(s))
        


```

```python
#w6a5
def count(nums):
    dt=dict()
    gr=0
    less=0
    equal=0
    for i in nums:
        if i>0: gr+=1
        elif i<0: less+=1
        else: equal+=1
    dt["positives"]=gr
    dt["negatives"]=less
    dt["zeros"]=equal
    return dt
nums=list(map(int,input().split()))
print(count(nums))
```

```python
#w6a8
def counting(strings):
    dt=dict()
    for i in strings:
        if dt.get(i):
           dt[i]=dt[i]+1
        else:
            dt[i]=1
    return dt
strings=list(map(str,input().split()))
print(counting(strings))
```

```python
#w6a9
def merge2dict(dict1st,dict2nd):
    copydict=dict1st
    for i in dict2nd:
        if (i in copydict):
            copydict[i]+=dict2nd[i]
        else:
            copydict[i]=dict2nd[i]
    s=[]
    lastdict={}
    for i in copydict:
        s.append(i)
    s=sorted(s)
    for i in s:
        lastdict[i]=copydict[i]
    return lastdict
    
n,m=(map(int,input().split()))
dict1st={}
for i in range(n):
    a=input()
    b=int(input())
    dict1st[a]=b
dict2nd={}
for i in range(m):
    a=input()
    b=int(input())
    dict2nd[a]=b
print(merge2dict(dict1st,dict2nd))
```

```python
def twosum(nums,target):
    n=len(nums)
    dt={}
    res=set()
    for i in range(n):
        val=target-nums[i]
        if(val in dt):
            tu=(val,nums[i])
            res.add(tu)
        dt[nums[i]]=i
    return list(res)
nums=list(map(int,input().split()))
target=int(input())
print(twosum(nums,target))
```

```python
#w6a12
def commonnums(nums):
    dt=dict()
    for i in nums:
        if i in dt:
            dt[i]+=1
        else:
            dt[i]=1
    most=1e9
    mostvalue=0
    for i in dt:
        if (dt[i]>mostvalue):
            mostvalue=dt[i]
            most=i
        elif(dt[i]==mostvalue):
            most=min(most,i)        
    return most
nums=list(map(int,input().split()))
print(commonnums(nums))            
```

```python
#w6a14
def inters(a,b):
    intersect=[]
    for i in a:
        if i in b and not i in intersect:
            intersect.append(i)
    return intersect
a=list(map(int,input().split()))
b=list(map(int,input().split()))
print(inters(a,b))
```

```python
#w6a15
def refine(s,target):
    s=s+" "
    n=len(s)
    dt={}
    j=0
    k=0
    x=True
    for i in range(n):
        if s[i]==' ':
            if x:
                k=i
                x=False
            else:
                if int(s[k+1:i])>target:
                    dt[s[j:k]]=int(s[k+1:i])
                x=True
                j=i+1
    return dt
s=input()
k=int(input())
print(refine(s,k))

                    
                
```

```python
#w6a19
def pos_of_max(nums):
    n=len(nums)
    max_pos=0
    for i in range(n):
        if (nums[i]>nums[max_pos]):
            max_pos=i
    return max_pos+1
nums=list(map(int,input().split()))
print(pos_of_max(nums))
```

```python
#w6a20
def find_k(nums,k):
    n=len(nums)
    for i in range(n):
        if(nums[i]==k):
            return i+1
    return -1
nums=list(map(int,input().split()))
k=int(input())
print(find_k(nums,k))
```

```python
#w6a21
def mapppp(s):
    s=s+" "
    n=len(s)
    dt={}
    j=0
    coun=0
    for i in range(n):
        if s[i]==' ':
            l=[]
            if s[j:i] in dt:
                l=dt[s[j:i]]
                l.append(coun)
                dt[s[j:i]]=l
            else:
                l.append(coun)
                dt[s[j:i]]=l
            j=i+1
            coun+=1
    for i in dt:
        dt[i]=tuple(dt[i])
    return dt
s=input()
print(mapppp(s))

```
